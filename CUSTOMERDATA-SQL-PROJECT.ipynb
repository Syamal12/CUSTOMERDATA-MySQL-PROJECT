```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandasql import sqldf

```

```python
#data to dataframe
DF2=pd.DataFrame(df)
```

```python
#head
Q="SELECT * FROM DF2 LIMIT 5"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#tail
Q="SELECT * FROM DF2 ORDER BY ROWID DESC LIMIT 5"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#data cleaning
Q="SELECT * FROM DF2 WHERE customer_id IS NULL"
RESULT=sqldf(Q,locals())
RESULT
```

```python
Q="SELECT * FROM DF2 WHERE gender IS NULL"
RESULT=sqldf(Q,locals())
RESULT
```

```python
Q="SELECT *FROM DF2 WHERE age IS NULL;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
Q="SELECT * FROM DF2 WHERE payment_method IS NULL;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Retrieve Customers Based on Specific Age
Q="SELECT * FROM DF2 WHERE age >68;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find Customers with a Specific Payment Method
Q="SELECT * FROM DF2 WHERE payment_method ='Credit Card';"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Count the Number of Customers in Each Payment Method Category
Q="SELECT payment_method, COUNT(*) AS customer_count FROM DF2 GROUP BY payment_method;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find Customers in a Specific Age Range
Q="SELECT * FROM DF2 WHERE age BETWEEN 65 AND 69;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Count the Number of Male and Female Customers
Q="SELECT gender, COUNT(*) AS customer_count FROM DF2 GROUP BY gender;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find the Oldest Customer
Q="SELECT * FROM DF2 ORDER BY age DESC LIMIT 1;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find the Youngest Customer
Q = "SELECT * FROM DF2 ORDER BY age ASC LIMIT 1;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find the Total Number of Customers (Count All Rows)
Q="SELECT COUNT(*) AS total_customer FROM DF2;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find customer_id and payment_method
Q="SELECT customer_id, payment_method FROM DF2;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find Customers Who Have Used Multiple Payment Methods
Q="SELECT customer_id FROM DF2 GROUP BY customer_id HAVING COUNT(DISTINCT payment_method) >1;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#List All Unique Payment Methods Used
Q="SELECT DISTINCT payment_method FROM DF2;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find Customers Who Have Used Only One Payment Method
Q="SELECT customer_id FROM DF2 GROUP BY customer_id HAVING COUNT(DISTINCT payment_method) =1;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find the Average Age of Customers
Q="SELECT AVG(age) AS average_age FROM DF2;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#List Customers Who Are Below 19 and Use Credit Card
Q="SELECT * FROM DF2 WHERE age<19 AND payment_method='Credit Card';"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Get the Customer with the Most Common Payment Method
Q="SELECT payment_method,COUNT(*) AS count FROM DF2 WHERE age BETWEEN 30 AND 40 GROUP BY payment_method ORDER BY count DESC LIMIT 1;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Rank Payment Methods Based on Usage
Q="SELECT payment_method, COUNT(*) AS transactionS,RANK() OVER(ORDER BY COUNT(*)DESC) AS RANK FROM DF2 GROUP BY payment_method;"
RESULT=sqldf(Q,locals())
RESULT
```

```python
#Find the Percentage of Male and Female Customers
Q="SELECT gender ,COUNT(*) * 100.0/ SUM(COUNT(*)) OVER() AS percentage FROM DF2 GROUP BY GENDER;"
RESULT=sqldf(Q,locals())
RESULT
```
